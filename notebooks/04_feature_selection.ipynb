```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway, chi2_contingency
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

# Phase 4: Statistical Testing & Feature Selection
Conducting statistical tests (SciPy) like t-tests/ANOVA to validate feature significance against the target variable, followed by dimensionality reduction (PCA) if necessary, to finalize the ML-ready dataset.

```python
feature_df  = pd.read_csv('../data/processed/engineered_market_data.csv')
```

```python
feature_df.head()
```

```python
selected_df = feature_df.copy()
```

```python
selected_df.head()
```

```python
selected_df.set_index('Date', inplace= True)
```

```python
X = selected_df.drop('Target', axis= 1)
```

```python
X.head()
```

```python
y = selected_df['Target']
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, shuffle= False)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
X_test.columns
```

```python
test_columns = ['Close', 'High', 'Low', 'Open', 'Volume', 'RSI', 'Doji','Bullish_Engulfing', 'Bearish_Engulfing', 'Hammer',
           'Ticker_Name_INFY.NS', 'Ticker_Name_ITC.NS', 'Ticker_Name_LT.NS','Ticker_Name_RELIANCE.NS', 'Ticker_Name_TCS.NS',
           'Ticker_Name_^NSEBANK','Ticker_Name_^NSEI']
```

```python
f_oneway_result: list = []
```

```python
significant_value = 0.05

for column in test_columns:
     group_0 = X_train[column][y_train == 0]
     group_1 = X_train[column][y_train == 1]
     f_stats_value , p_value = f_oneway(group_0, group_1)
     result = 'H0 REJECT' if p_value < significant_value else 'H0 FAIL TO REJECT'
     f_oneway_test: dict = {
          'feature'           : column,
          'f_stats_value'     : f_stats_value,
          'p_value'           : p_value,
          'result'            : result
     }

     f_oneway_result.append(f_oneway_test)
```

```python
f_oneway_result_df = pd.DataFrame(f_oneway_result)
```

```python
f_oneway_result_df
```

```python
rf_model = RandomForestClassifier(random_state= 42)
```

```python
rf_model.fit(X_train, y_train)
```

```python
feature_importances = rf_model.feature_importances_
feature_importances
```

```python
feature_importances_df = pd.DataFrame(feature_importances.reshape(1,21), columns= X_train.columns)
feature_importances_df
```
